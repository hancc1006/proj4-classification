```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_val_score, cross_val_predict
from sklearn.model_selection import (cross_val_score, train_test_split, 
                                     KFold, GridSearchCV)
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold, SelectFromModel
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from scipy import stats
```

```python
le=preprocessing.LabelEncoder()
```

```python
hp_df1['Encoded_Rural_Code']=le.fit_transform(hp_df1['Rural_Status_Code'])
hp_df1['Encoded_US_Mexico_Border_100km']=le.fit_transform(hp_df1['US_Mexico_Border_100km'])
```

```python
hp_df1['Poverty_Level']=le.fit_transform(hp_df1['HPSA_Score']<15)
```

```python
list_features1=['Designation_Pop', 'Estimated_Underserved_Pop', 'Estimated_Served_Pop',
       'Total_FTE_Clinicians','Percent_Pop_Below_Poverty', 'HPSA_Shortage',
       'CountyFIPS', 'Geocoding_Primary_X', 'Geocoding_Primary_Y', 
        'Year', 'Population','Encoded_Rural_Code','Encoded_US_Mexico_Border_100km', 'HPSA_Score']
```

```python
hp_df2=hp_df1.fillna(0)
```

```python
X=hp_df2[list_features1]
#y=hp_df2['HPSA_Score']
y=hp_df2['Poverty_Level']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
hp1=LogisticRegression(C=1000)
hp1.fit(X_train,y_train)
y_log_pred=hp1.predict(X_test)
score=hp1.score(X_test,y_test)
accuracy=cross_val_score(hp1,X,y,cv=10,scoring="r2").mean()
print(score, accuracy)
print(f"Precision:{precision_score(y_test,y_log_pred,average='macro')}")
print(f"Recall:{recall_score(y_test,y_log_pred,average='macro')}")
print(f"f1:{f1_score(y_test,y_log_pred,average='macro')}")
```

```python
log_confusion=confusion_matrix(y_test, y_log_pred)
plt.figure(dpi=150)
sns.heatmap(log_confusion,cmap=plt.cm.Blues, square=True)
plt.xlabel('Predicted Score')
plt.ylabel('Actual Score')
plt.title('Logistic Regression Confusion Matrix')
plt.tight_layout()
plt.savefig('Logistic Regression Confusion Matrix.jpg',bbox_inches='tight',dpi=300)
```

```python
neigh=KNeighborsClassifier()
neigh.fit(X_train,y_train)
knn_pred=neigh.predict(X_test)
score=neigh.score(X_test,y_test)
accuracy=cross_val_score(neigh,X,y,cv=10,scoring="r2").mean()
print(score, accuracy)
print(f"Precision:{precision_score(y_test,knn_pred,average='macro')}")
print(f"Recall:{recall_score(y_test,knn_pred,average='macro')}")
print(f"f1:{f1_score(y_test,knn_pred,average='macro')}")
```

```python
knn_pred=neigh.predict_proba(X_test)
score=neigh.score(X_test,y_test)
accuracy=cross_val_score(neigh,X,y,cv=10,scoring="r2").mean()
print(score, accuracy)
#print(f"Precision:{precision_score(y_test,knn_pred)}")
#print(f"Recall:{recall_score(y_test,knn_pred)}")
#print(f"f1:{f1_score(y_test,knn_pred)}")
```

```python
knn_confusion=confusion_matrix(y_test, knn_pred)
plt.figure(dpi=150)
sns.heatmap(knn_confusion,cmap=plt.cm.Blues, square=True)
plt.xlabel('Predicted Score')
plt.ylabel('Actual Score')
plt.title('KNN Confusion Matrix')
plt.savefig('KNN Confusion Matrix.jpg',bbox_inches='tight',dpi=300)
```

```python
#from module2 week 3
threshold=0.5
rf = RandomForestRegressor(n_estimators=2000, max_features=3, max_depth=5)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
y_pred
```

```python
y_pred=np.where(y_pred>threshold,0,1)
print("Random Forest Regression:", r2_score(y_test, y_pred))
print(f"Precision:{precision_score(y_test,y_pred,average='macro')}")
print(f"Recall:{recall_score(y_test,y_pred,average='macro')}")
print(f"f1:{f1_score(y_test,y_pred,average='macro')}")
```

```python
#from module2 week 3
gbm = GradientBoostingRegressor(n_estimators=500, max_depth=3, learning_rate=.01)
gbm.fit(X_train, y_train)
y_pred = gbm.predict(X_test)
y_pred=np.where(y_pred>threshold,0,1)
print("Gradient Boosted Regression:", r2_score(y_test, y_pred))
print(f"Precision:{precision_score(y_test,y_pred,average='macro')}")
print(f"Recall:{recall_score(y_test,y_pred,average='macro')}")
print(f"f1:{f1_score(y_test,y_pred,average='macro')}")
```
